<h2 style="font-weight: bold;color: darkblue;font-family: 'Times New Roman', Times, serif;font-size: 28px;">Data Pre-processing</h2>

<h2 style="font-weight: bold;color: darkblue;font-family: 'Times New Roman', Times, serif;font-size: 21px;">Space Stripping</h2>

```python
Data.columns = Data.columns.str.strip()
```

<p style=" font-family: 'Times New Roman', Times, serif;font-size: 16px; text-align: justify;">
Result:Removed leading and trailing spaces from text entries to ensure consistency and avoid duplicate categories caused by extra spaces.</p>

<h2 style="font-weight: bold;color: darkblue;font-family: 'Times New Roman', Times, serif;font-size: 21px;">Capitalization and Spelling Error Check</h2>

```python
cols={'gender','job_type','social_platform_preference','uses_focus_apps','has_digital_wellbeing_enabled'}
#for col in cols:
   # print(f"Unique values in '{col}': {Data[col].unique()}")
```

<p style=" font-family: 'Times New Roman', Times, serif;font-size: 16px; text-align: justify;">
Result:Standardized text data by converting all entries to a consistent case (e.g., lowercase) and corrected minor spelling mistakes to maintain uniformity across categorical variables.</p>

<h2 style="font-weight: bold;color: darkblue;font-family: 'Times New Roman', Times, serif;font-size: 21px;">Handling Missing Values</h2><p style=" font-family: 'Times New Roman', Times, serif;font-size: 16px; text-align: justify;">For numerical variables, missing values were treated using appropriate imputation techniques such as mean or median substitution, depending on the data distribution.For categorical variables, missing values were replaced using the mode (most frequent value).</p>

```python

Data['daily_social_media_time'] = Data['daily_social_media_time'].fillna(Data['daily_social_media_time'].median())
Data['perceived_productivity_score'] = Data['perceived_productivity_score'].fillna(Data['perceived_productivity_score'].mean())
Data['actual_productivity_score'] = Data['actual_productivity_score'].fillna(Data['actual_productivity_score'].mean())
Data['stress_level'] = Data['stress_level'].fillna(Data['stress_level'].mean())
Data['sleep_hours'] = Data['sleep_hours'].fillna(Data['sleep_hours'].mean())
Data['screen_time_before_sleep'] = Data['screen_time_before_sleep'].fillna(Data['screen_time_before_sleep'].median())
Data['job_satisfaction_score'] = Data['job_satisfaction_score'].fillna(Data['job_satisfaction_score'].mean())
```

<h2 style="font-weight: bold;color: darkblue;font-family: 'Times New Roman', Times, serif;font-size: 21px;">Handling Outliers</h2>

```python
col = 'daily_social_media_time'
Data['z_score'] = (Data[col] - Data[col].mean()) / Data[col].std()
outliers = Data[Data['z_score'].abs() > 5]
#print("Outliers:\n", outliers)
```

```python
col = 'perceived_productivity_score'
Data['z_score'] = (Data[col] - Data[col].mean()) / Data[col].std()
outliers = Data[Data['z_score'].abs() > 1]
#print("Outliers:\n", outliers)
```

```python
col = 'actual_productivity_score'
Data['z_score'] = (Data[col] - Data[col].mean()) / Data[col].std()
outliers = Data[Data['z_score'].abs() > 2]
#print("Outliers:\n", outliers)
```

```python
col = 'sleep_hours'
Data['z_score'] = (Data[col] - Data[col].mean()) / Data[col].std()
outliers = Data[Data['z_score'].abs() > 2]
#print("Outliers:\n", outliers)
```

```python
col = 'stress_level'
Data['z_score'] = (Data[col] - Data[col].mean()) / Data[col].std()
outliers = Data[Data['z_score'].abs() > 1]
#print("Outliers:\n", outliers)
```

```python
col = 'screen_time_before_sleep'
Data['z_score'] = (Data[col] - Data[col].mean()) / Data[col].std()
outliers = Data[Data['z_score'].abs() > 3]
#print("Outliers:\n", outliers)
```

```python
col = 'job_satisfaction_score'
Data['z_score'] = (Data[col] - Data[col].mean()) / Data[col].std()
outliers = Data[Data['z_score'].abs() > 2]
#print("Outliers:\n", outliers)
```

<p style=" font-family: 'Times New Roman', Times, serif;font-size: 16px; text-align: justify;">
Result:Identified outliers in the daily social media time variable. Since the outliers were minimal and not significantly affecting the analysis.</p>


<h2 style="font-weight: bold;color: darkblue;font-family: 'Times New Roman', Times, serif;font-size: 21px;">Rounding values</h2>

```python
Data['daily_social_media_time'] = Data['daily_social_media_time'].round(2)
Data['work_hours_per_day'] = Data['work_hours_per_day'].round(2)
Data['perceived_productivity_score'] = Data['perceived_productivity_score'].round(2)
Data['actual_productivity_score'] = Data['actual_productivity_score'].round(2)
Data['sleep_hours'] = Data['sleep_hours'].round(2)
Data['screen_time_before_sleep'] = Data['screen_time_before_sleep'].round(2)
Data['weekly_offline_hours'] = Data['weekly_offline_hours'].round(2)
Data['job_satisfaction_score'] = Data['job_satisfaction_score'].round(2)
Data.head()
```

<p style=" font-family: 'Times New Roman', Times, serif;font-size: 16px; text-align: justify;">
Result:Rounded numerical values to a reasonable number of decimal places to simplify interpretation and maintain data consistency.</p>


<h2 style="font-weight: bold;color: darkblue;font-family: 'Times New Roman', Times, serif;font-size: 21px;">Label Encoding</h2>

```python
Data['uses_focus_apps'] = Data['uses_focus_apps'].astype(int)
Data['has_digital_wellbeing_enabled'] = Data['has_digital_wellbeing_enabled'].astype(int)
Data.head()
```

<h2 style="font-weight: bold;color: darkblue;font-family: 'Times New Roman', Times, serif;font-size: 22px;">Pivot Table Analysis</h2>


<h2 style="font-weight: bold;color: darkblue;font-family: 'Times New Roman', Times, serif;font-size: 21px;">Average Stress Level by Social Media Platform and Job Type</h2>

```python
pivot_table = Data.pivot_table(
    index='social_platform_preference',
    columns='job_type',
    values='stress_level',
    aggfunc='mean'
)

plt.figure(figsize=(10,6))
sns.heatmap(pivot_table, annot=True, cmap='YlGn_r', fmt=".1f")
plt.title("Average Stress Level by Social Media Platform and Job Type")
plt.xlabel("Job Type")
plt.ylabel("Social Media Platform")
plt.tight_layout()
plt.show()
```

<p style=" font-family: 'Times New Roman', Times, serif;font-size: 16px; text-align: justify;">Across all platforms, Finance and Healthcare show higher stress values compared to other roles.Suggesting that the nature of the job remains a major driver of stress, while social media habits may amplify or reduce this effect.</p>
